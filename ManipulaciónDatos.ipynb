### Filtrando el DataFrame

### Agrupar datos

```python
#agrupación básico con una variable y su agregación
df_group = df.groupby('ShipName')['Quantity'].mean()
df_group
```

```python
#diferentes funciones de agregación  median(), value_counts(normalize = True)
df_group1 = df.groupby('ShipName')['Quantity'].agg([min,max,sum,np.mean])
df_group1
```

### Tablas dinámicas

```python
#agregar una nueva columna calculada
df['SalesAmount']= df['Quantity'] *  df['UnitPrice']

#creando la tabla dinámica
pivot_table = df.pivot_table(
    'SalesAmount', index= 'ShipCountry', columns= 'ShipperName',
    aggfunc=[np.median,np.sum])
pivot_table
```

```python
#rellenar los valores vacíos y agregar los totales
pivot_table1 = df.pivot_table(
    'SalesAmount', index= 'ShipCountry', columns= 'ShipperName',
    aggfunc=[np.median,np.sum],fill_value=0, margins= True)
pivot_table1
```

```python

dft= pd.read_csv(r'C:\Users\Outlet\Desktop\Nueva carpeta\BDTienda.csv',sep=',',header=0, index_col=False, encoding='latin-1',engine='python')
dft
```

```python
#especificar columna como fecha

dft['Order Date'] = pd.to_datetime(dft['Order Date'], dayfirst=True, errors='coerce')
dft['Ship Date'] = pd.to_datetime(dft['Ship Date'], dayfirst=True, errors='coerce')

print(dft[['Order Date', 'Ship Date']].head())
```

```python
#personalizar formato de fecha
dft['Order_date'] = dft['Order Date'].dt.strftime('%d-%m-%Y')
dft
```

```python
#agrupar por fecha
dft.resample('ME', on='Ship Date')['Sales'].mean()
```

```python
#extrayendo datos a partir de fechas
#nombre del día
dft['dia'] = pd.to_datetime(dft['Ship Date']).dt.day_name()

#año
dft['anio'] = pd.to_datetime(dft['Ship Date']).dt.year

#mes
dft['mes'] = pd.to_datetime(dft['Ship Date']).dt.month_name()

dft
```
